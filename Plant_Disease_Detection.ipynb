```python
import os
import shutil
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# --- STEP 1: DOWNLOAD AND ORGANIZE THE DATASET ---

# If you haven't already, run the Kaggle download commands first
# !pip install kaggle
# from google.colab import files
# files.upload() # Upload kaggle.json
# !mkdir ~/.kaggle
# !mv kaggle.json ~/.kaggle/
# !chmod 600 ~/.kaggle/kaggle.json
# !kaggle datasets download -d abdallahalidev/plantvillage-dataset

# Unzip the dataset
!unzip plantvillage-dataset.zip -d /content/plant_data_unzipped


# --- IMPORTANT: CORRECT THE PATH ---
# The downloaded dataset is likely nested within a few folders.
# Let's find the correct path to the class folders.
# It is most likely inside 'plant_data_unzipped/plantvillage-dataset/color'.
main_dir = '/content/plant_data_unzipped/plantvillage dataset/color'

# Create the new organized directories
base_dir_organized = '/content/plant_data_organized'
train_dir = os.path.join(base_dir_organized, 'train')
validation_dir = os.path.join(base_dir_organized, 'validation')

os.makedirs(train_dir, exist_ok=True)
os.makedirs(validation_dir, exist_ok=True)

# Loop through each class folder and split the images
try:
    for class_folder in os.listdir(main_dir):
        class_path = os.path.join(main_dir, class_folder)

        # Check if it's a directory
        if os.path.isdir(class_path):
            all_files = os.listdir(class_path)
            train_files, val_files = train_test_split(all_files, test_size=0.2, random_state=42)

            # Create class directories in the new train and validation folders
            os.makedirs(os.path.join(train_dir, class_folder), exist_ok=True)
            os.makedirs(os.path.join(validation_dir, class_folder), exist_ok=True)

            # Copy files to the new train directory
            for file_name in train_files:
                src = os.path.join(class_path, file_name)
                dst = os.path.join(train_dir, class_folder, file_name)
                shutil.copyfile(src, dst)

            # Copy files to the new validation directory
            for file_name in val_files:
                src = os.path.join(class_path, file_name)
                dst = os.path.join(validation_dir, class_folder, file_name)
                shutil.copyfile(src, dst)

    print("Dataset successfully organized.")

except FileNotFoundError:
    print(f"ERROR: Could not find the directory: {main_dir}. Please check your file path.")
    print("The most likely issue is the folder name. Please inspect the unzipped files in the Colab file browser and adjust the 'main_dir' variable accordingly.")
    raise # Re-raise the exception to stop execution


# --- STEP 2: DATA PREPROCESSING AND AUGMENTATION ---
# This part of the code is correct and will work once the file path is fixed.

IMAGE_SIZE = (128, 128)
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255, rotation_range=20, width_shift_range=0.2,
    height_shift_range=0.2, shear_range=0.2, zoom_range=0.2,
    horizontal_flip=True, fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

class_labels = list(train_generator.class_indices.keys())
num_classes = len(class_labels)
print(f"Number of classes: {num_classes}")
print(f"Class labels: {class_labels}")


# --- STEP 3: BUILD AND TRAIN THE CNN MODEL ---

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE
)


# --- STEP 4: EVALUATE AND VISUALIZE RESULTS ---

loss, accuracy = model.evaluate(validation_generator)
print(f"Validation Loss: {loss:.4f}")
print(f"Validation Accuracy: {accuracy:.4f}")

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(len(acc))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
# Save the trained model to a file
model.save('plant_disease_model.h5')
print("Model saved as plant_disease_model.h5")
```

```python
from google.colab import files
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os

# Define the descriptions for each disease
# You can expand this dictionary with more details for all 38 classes
disease_descriptions = {
    "Apple___Black_rot": "Black rot is a fungal disease caused by Botryosphaeria obtusa. It creates black, rotting spots on fruit and 'frogeye' lesions on leaves.",
    "Apple___healthy": "The plant is healthy with no visible signs of disease.",
    "Corn_(maize)___Common_rust": "Common rust is a fungal disease caused by Puccinia sorghi, identified by cinnamon-brown pustules on leaves.",
    "Grape___Esca_(Black_Measles)": "Esca, or black measles, is a fungal disease that causes leaf discoloration, often in a 'tiger stripe' pattern, and spotting on the fruit.",
    "Tomato___Bacterial_spot": "Bacterial spot is a disease caused by four species of Xanthomonas bacteria. It creates small, circular, water-soaked spots on leaves and scabby lesions on fruit.",
    "Tomato___Early_blight": "Early blight is a fungal disease caused by Alternaria solani. It produces target-like spots with concentric rings, primarily on lower leaves.",
    "Potato___Early_blight": "Potato early blight is a fungal disease caused by Alternaria solani. It causes small, dark lesions on older leaves, often surrounded by a yellow halo.",
    "Pepper,_bell___Bacterial_spot": "Pepper bacterial spot is caused by Xanthomonas campestris pv. vesicatoria. It leads to dark, circular spots with yellow halos on leaves and raised spots on fruit.",
    "Grape___Leaf_blight_(Isariopsis_Leaf_Spot)": "Leaf blight is a fungal disease caused by Pseudocercospora vitis. It causes irregular, brownish-red spots with dark borders on the leaf surface.",
    "Peach___Bacterial_spot": "Peach bacterial spot is caused by Xanthomonas arboricola. It creates small, dark spots with yellow halos on leaves, often leading to shot-hole symptoms.",
}

# Function to get the description for a predicted disease
def get_description(disease_name):
    return disease_descriptions.get(disease_name, "Description not available for this disease.")

# Step 1: Upload a test image from your computer
print("Please upload a test image of a plant leaf.")
uploaded = files.upload()

# Get the name of the uploaded file
uploaded_file_name = next(iter(uploaded))

# Step 2: Load the saved model
try:
    loaded_model = load_model('plant_disease_model.h5')
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")
    print("Please make sure you have run the model.save() command and the model file 'plant_disease_model.h5' exists.")
    exit()

# Step 3: Prepare the uploaded image for prediction
img = image.load_img(uploaded_file_name, target_size=(128, 128))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0

# Step 4: Make a prediction
predictions = loaded_model.predict(img_array)
class_labels = list(train_generator.class_indices.keys())

# Step 5: Visualize the results
plt.figure(figsize=(15, 6))

# Subplot 1: Display the uploaded image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Uploaded Image")
plt.axis('off')

# Subplot 2: Display the predictions as a bar chart
plt.subplot(1, 2, 2)
# Get the top 5 predictions
top_5_indices = np.argsort(predictions[0])[::-1][:5]
top_5_probabilities = predictions[0][top_5_indices]
top_5_labels = [class_labels[i] for i in top_5_indices]

plt.barh(top_5_labels, top_5_probabilities, color='skyblue')
plt.title("Prediction Confidence")
plt.xlabel("Confidence")
plt.ylabel("Disease")
plt.xlim(0, 1.0)
plt.gca().invert_yaxis()  # Display the highest confidence at the top

plt.tight_layout()
plt.show()

# Step 6: Get the top prediction and its description
predicted_class_index = np.argmax(predictions[0])
predicted_label = class_labels[predicted_class_index]
confidence = np.max(predictions[0]) * 100
disease_info = get_description(predicted_label)

# Step 7: Print the final report
print("\n--- Model Prediction Report ---")
print(f"Prediction: {predicted_label}")
print(f"Confidence: {confidence:.2f}%")
print("\n--- Disease Information ---")
print(f"Description: {disease_info}")
```
